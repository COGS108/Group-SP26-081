## Background and Prior Work


From 2010 to 2020, the PC gaming landscape underwent a major transformation as AAA games increased in scale, graphical fidelity, and simulation complexity. These advances led developers to steadily raise minimum system requirements, requiring more powerful CPUs, GPUs, and larger amounts of RAM. However, this rapid increase in software demands did not perfectly align with consumer hardware adoption. While developers pushed toward higher performance baselines, many PC gamers continued using mid-range systems for extended periods, creating a growing gap between required and available hardware.

A key factor behind rising requirements was the transition toward multi-core CPU utilization in modern game engines. Earlier titles primarily relied on single-core performance, but by the early 2010s, developers began optimizing engines to take advantage of multiple cores for tasks such as physics simulation, artificial intelligence, and world streaming. This shift increased the baseline expectation for CPUs, as games increasingly depended on parallel processing rather than raw clock speed. At the same time, improvements in console hardware also influenced PC requirements, as cross-platform development pushed developers to target higher-performance systems overall<a name="ref1"></a><a href="#fn1">1</a>.

In parallel, GPU requirements increased significantly due to advancements in rendering techniques. Higher-resolution textures, dynamic lighting, and more complex shader systems placed greater strain on graphics hardware. <a name="ref2"></a>Historical analyses of GPU performance and gaming requirements show a clear upward trend in memory and processing demands, with modern AAA titles often requiring several gigabytes of VRAM and significantly more processing power than earlier games<a href="#fn2">2</a>. These graphical improvements were a major driver of hardware upgrades throughout the latter half of the decade.

Despite these increasing requirements, data from the Steam Hardware Survey demonstrates that consumer hardware adoption progressed more gradually. The survey, which collects hardware data from millions of users, is widely used by developers to understand the distribution of real-world PC configurations<a name="ref3"></a><a href="#fn3">3</a>. Even in recent years, mid-range GPUs and moderate system configurations have remained the most common among players, rather than the high-end hardware typically recommended for modern AAA titles. This highlights a persistent mismatch between developer expectations and the hardware that most players actually own, reinforcing the accessibility challenges created by rising system requirements.

Some prior work that we found was an article posted on Windows Central by Cale Hunt. He analyzed the hardware regression from February 2026 to March 2026, and noticed a drastic decrease in hardware quality. He did this by analyzing the Steam Hardware Survey for those two months, and cross referenced that data with hardware pricing from Spring 2025, to Spring 2026. For the average user the RAM decreased, storage decreased, and the use of higher resolution monitors decreased.<a name="ref4" href="#fn4">4</a> What attributed to this was the volatile pricing at the time. In that time period, RAM prices were up drastically, with storage prices increasing too. With these drastic price hikes, users have to settle on lesser specs, which explains the decrease in hardware.

### Steam Hardware Survey Dataset (2010–2020)

- Dataset Name: Steam Hardware Survey Dataset (shs.csv)
- Link to dataset: https://github.com/jdegene/steamHWsurvey/blob/master/shs.csv
- Number of observations: Approximately 66,041 observations before filtering and about 11,325 observations after cleaning and restricting the dataset to relevant years and hardware categories.
- Number of variables: 5 original variables: ``[date, category, name, change, percentage]``, & another derived variable: ``[year]``
- Description of variables most relevant: 
``date`` records when the survey data was collected.
``category`` identifies the hardware type being measured, such as CPU speed, RAM, or GPU.
``name`` contains the specific hardware specification or model.
``percentage`` represents the proportion of Steam users using that hardware configuration during a given month.
``year`` was created from the date column to simplify yearly trend analysis.

This dataset contains monthly hardware survey information collected from Steam users between 2010 and 2020. The Steam Hardware Survey is conducted by Valve and records information about the computer hardware used by players on the Steam platform. For this project, the dataset was filtered to focus specifically on hardware categories relevant to gaming performance: CPU speed, system RAM, and video card descriptions. These variables are important because the project aims to compare the minimum hardware requirements of AAA PC games with the hardware commonly owned by average Steam users over time.

The dataset has been cleaned and tidied. The date column was converted into a datetime format and a separate year column was created to simplify yearly comparisons. The dataset was then filtered to only include observations from 2010 through 2020 to match the scope of the research question. Categories unrelated to gaming hardware were removed, leaving only CPU, RAM, and GPU information. Missing values were inspected and rows missing hardware names were removed because they could not be meaningfully analyzed. Entries with invalid percentage values outside the range of 0 to 1 were also removed. In addition, generic GPU labels such as “Other” were excluded because they were too broad to provide useful information about actual consumer hardware trends.


The biggest concern that comes with ths dataset is that the Steam Hardware Survey only represents Steam users and not the entire population of PC gamers, this could introduce some sampling bias because Steam users may differ from other PC gamers. Another issue that this dataset represents is that GPU performance is difficult to measure directly from categorical data of model names alone, comparing GPUs numerically is more complicated than say comparing CPU speeds that have a quantifiable measure.

```python
import pandas as pd

Steam_HW_Survey = pd.read_csv('data/00-raw/shs.csv')

print("Original dataset shape:")
print(Steam_HW_Survey.shape)

Steam_HW_Survey['date'] = pd.to_datetime(Steam_HW_Survey['date'])

Steam_HW_Survey = Steam_HW_Survey[
    (Steam_HW_Survey['date'].dt.year >= 2010) &
    (Steam_HW_Survey['date'].dt.year <= 2020)
]

categories = [
    'AMD CPU Speeds',
    'Intel CPU Speeds',
    'System RAM',
    'Video Card Description'
]

Steam_HW_Survey = Steam_HW_Survey[Steam_HW_Survey['category'].isin(categories)]

Steam_HW_Survey['year'] = Steam_HW_Survey['date'].dt.year

print("\nRelevant categories:")
print(Steam_HW_Survey['category'].unique())

print("\nMissing values:")
print(Steam_HW_Survey.isnull().sum())

print("\nRows with missing values:")
print(Steam_HW_Survey[Steam_HW_Survey.isnull().any(axis=1)].shape[0])

print("\nSuspicious percentage values:")
print(Steam_HW_Survey[(Steam_HW_Survey['percentage'] < 0) | (Steam_HW_Survey['percentage'] > 1)].shape[0])

Steam_HW_Survey = Steam_HW_Survey[(Steam_HW_Survey['percentage'] >= 0) & (Steam_HW_Survey['percentage'] <= 1)]

Steam_HW_Survey = Steam_HW_Survey.dropna(subset=['name'])

Steam_HW_Survey['category'] = Steam_HW_Survey['category'].str.strip()
Steam_HW_Survey['name'] = Steam_HW_Survey['name'].str.strip()

Steam_HW_Survey = Steam_HW_Survey[Steam_HW_Survey['name'] != 'Other']

print("\nCleaned dataset shape:")
print(Steam_HW_Survey.shape)

print("\nCleaned dataset preview:")
display(Steam_HW_Survey.head())

print("\nMost common hardware by category and year:")

top_hardware = Steam_HW_Survey.loc[
    Steam_HW_Survey.groupby(['year', 'category'])['percentage'].idxmax()
]

top_hardware = top_hardware[
    ['year', 'category', 'name', 'percentage']
].sort_values(['year', 'category'])

display(top_hardware.head(20))

Steam_HW_Survey.to_csv('data/02-processed/cleaned_shs.csv', index=False)
```
